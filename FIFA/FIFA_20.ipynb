```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
data= pd.read_csv('players_20.csv')
data.info()
```

```python
k= DataFrame(data.columns)
k.tail(60)
```

```python
pot =data['potential'].groupby(data['club']).sum().reset_index()

pot_rank= pot.sort_values(by = 'potential', ascending = False)
club_potential = pot_rank.head(12)
```

```python
club_potential = DataFrame(club_potential)
club_potential
```

```python
import seaborn as sns
```

```python
club_potential.info()
```

```python
plt.figure(figsize=(17,10))
plt.figure(figsize=(17,10))

plt.title("Total potential of players in a club", fontsize = 25)
k=sns.barplot(x = 'club', y = 'potential', data = club_potential)
k.set(ylim=(2650,2900))
```
